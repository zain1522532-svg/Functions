```python
# Function to calculate grade
def calculate_grade(marks):
    if marks >= 90:
        grade = "A+"
    elif marks >= 80:
        grade = "A"
    elif marks >= 70:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    elif marks >= 50:
        grade = "D"
    else:
        grade = "F"
    return grade


# Main program
name = input("Enter student name: ")
marks = float(input("Enter marks (out of 100): "))

grade = calculate_grade(marks)

print("\n--- Student Report ---")
print("Name:", name)
print("Marks:", marks)
print("Grade:", grade)
```
